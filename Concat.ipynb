```python
participants = ['S','T']
for participant in participants:
    path = f'{data_path}/{participant}/'
    for i in range (8):
        #Convert mp4 to JSON
        str01 = f'{path}V{i + 1}.mp4'
        str02 = f'{path}{participant}T{i + 1}.json'
        print(str01)
        print(str02)
        
        os.system(f'mp4tojson {str01} > {str02}')

        #import biometric signals and video features
        bio_df = pd.read_csv(f'{path}{participant}T{i + 1}.csv')

        #import video file
        video_jsn = json.load(open(f'{path}{participant}T{i + 1}.json'))
        #video to dictionary
        video_dict = dict(video_jsn[2])
        #append data to dictionary
        dic = {}
        dic.update({'video': video_dict})

        for feature in bio_features:
            dic.update({feature : bio_df[[feature]]})
        with open(f"{path}T{i + 1}.pkl", "wb") as handle:
            pickle.dump(dic, handle, protocol = pickle.HIGHEST_PROTOCOL)
        print('task:', i + 1, dic.keys())
```
